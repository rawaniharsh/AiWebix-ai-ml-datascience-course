## ğŸ§  What is a Closure?

A **Closure** is a function that:

- **Remembers the variables** from the outer function
- Even **after the outer function has finished**

In simple terms:
> A function carrying its **environment** (variables) along with it.

---

## ğŸ§ª Real-World Analogy:

Imagine your mom packed a **tiffin** for you.

- Sheâ€™s not there in school, but her **tiffin (function)** still has the **food (data)** she gave.
- Wherever you go, you still have access to what she packed.

## ğŸ“˜ Why Do We Need Closures?

- To create **customised functions** (like a counter, or logger)
- To maintain **state** without using global variables or classes
- Common in **decorators, logging, callback systems**

````python
# ğŸ” LESSON: CLOSURES IN PYTHON

# âœ… STEP 1: Basic Nested Function (No Closure Yet)
def outer_function():
    message = "Hello from outer!"

    def inner_function():
        print(message)

    return inner_function

# outer_function() returns inner_function
# But inner_function still remembers 'message' even when outer is done
my_func = outer_function()
my_func()  # Output: Hello from outer!

# âœ… This is a Closure
# The inner function remembers 'message' from outer_function's environment

# ğŸ”„ Another Example: Personalised Greeter
def greet(name):
    def welcome():
        print(f"ğŸ‘‹ Hello, {name}! Welcome to AiWebix.")
    return welcome

greet_harsh = greet("Harsh")
greet_riya = greet("Riya")

greet_harsh()  # ğŸ‘‹ Hello, Harsh! Welcome to AiWebix.
greet_riya()   # ğŸ‘‹ Hello, Riya! Welcome to AiWebix.

# âœ… Each returned function remembers the 'name' passed earlier
# These are closures because they retain access to outer scope

# ğŸ’¡ Real-World Use Case: Counter using Closure
def make_counter():
    count = 0  # outer variable

    def counter():
        nonlocal count  # tells Python to use 'count' from outer scope
        count += 1
        return count

    return counter

counter1 = make_counter()
print(counter1())  # 1
print(counter1())  # 2
print(counter1())  # 3

# âœ… Here, each time we call counter1(), it remembers and updates 'count'

# ğŸ§  Another Use Case: Multiplier Factory
def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))  # 10
print(triple(4))  # 12

# âš ï¸ Common Mistake 1: Forgetting to return inner function
def wrong_closure():
    msg = "Oops!"
    def inside():
        print(msg)
    # return inside is missing

# Now you can't call inside() outside
oops = wrong_closure()
# oops()  âŒ will give error: 'NoneType' object is not callable

# âš ï¸ Common Mistake 2: Trying to modify without nonlocal
def bad_counter():
    count = 0
    def counter():
        count += 1  # âŒ This will raise UnboundLocalError
        return count
    return counter

# bad = bad_counter()
# print(bad())  âŒ Uncommenting this will crash the program

# âœ… Best Practice: Always use nonlocal when modifying outer scope variables

# ğŸ“ Mini Task:
# Create a closure that adds a fixed number to any input

def make_adder(fixed_number):
    def adder(x):
        return x + fixed_number
    return adder

add_10 = make_adder(10)
print(add_10(25))  # 35

add_50 = make_adder(50)
print(add_50(100))  # 150
````
