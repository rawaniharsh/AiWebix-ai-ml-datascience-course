## Polymorphism means â€” a single function or method name can behave differently depending on the object calling it.

### ğŸ§Real-World Example 1: "Drive" Method in Vehicles

Class: Vehicle (Base)
â†“
Car, Bike, Truck inherit from Vehicle

All have a method: .drive()

But each one behaves differently:
- Car.drive() â†’ "Driving a car with 4 wheels"
- Bike.drive() â†’ "Riding a bike with 2 wheels"
- Truck.drive() â†’ "Driving a truck with heavy load"


### ğŸ’³ Real-World Example 2: Payment System

Class: Payment (Base)
â†“
CreditCard, PayPal, UPI inherit from Payment

All have a method: .make_payment(amount)

But each one works differently:
- CreditCard.make_payment(100) â†’ processes via bank gateway
- PayPal.make_payment(100) â†’ uses PayPal API
- UPI.make_payment(100) â†’ verifies UPI PIN

```python
# ğŸ” Lesson: Polymorphism in Python

# âœ… Example 1: Polymorphism using method overriding with inheritance

class Vehicle:
    def drive(self):
        print("This vehicle is moving...")

class Car(Vehicle):
    def drive(self):
        print("ğŸš— Car is driving on the road.")

class Bike(Vehicle):
    def drive(self):
        print("ğŸï¸ Bike is speeding on the highway.")

class Truck(Vehicle):
    def drive(self):
        print("ğŸšš Truck is hauling heavy cargo.")

# ğŸ‘‡ All these objects share the same method name, but behave differently
vehicles = [Car(), Bike(), Truck()]

print("ğŸ”¸ Vehicle Polymorphism Example:")
for v in vehicles:
    v.drive()  # Python automatically calls the correct version of drive()

print("\n")


# âœ… Example 2: Built-in Polymorphism
print("ğŸ”¸ Built-in Function Polymorphism:")
print(len("Harsh"))         # len() on string
print(len([1, 2, 3, 4]))     # len() on list
print(len({1: "a", 2: "b"})) # len() on dictionary

# All use `len()` but it behaves based on the type of object

print("\n")


# âœ… Example 3: Duck Typing (very Pythonic)
# If it walks like a duck and quacks like a duck... it's a duck ğŸ¦†

class Cat:
    def speak(self):
        print("Meow...")

class Dog:
    def speak(self):
        print("Woof!")

def pet_speak(pet):
    # No type check! Python just calls the speak() method if it exists
    pet.speak()

print("ğŸ”¸ Duck Typing Example:")
dog = Dog()
cat = Cat()

pet_speak(dog)
pet_speak(cat)

print("\n")


# âœ… Example 4: E-commerce Payment System â€” Method Overriding

class Payment:
    def make_payment(self, amount):
        print(f"Processing generic payment of â‚¹{amount}")

class CreditCard(Payment):
    def make_payment(self, amount):
        print(f"ğŸ’³ Paid â‚¹{amount} using Credit Card.")

class PayPal(Payment):
    def make_payment(self, amount):
        print(f"ğŸ’² Paid â‚¹{amount} using PayPal account.")

class UPI(Payment):
    def make_payment(self, amount):
        print(f"ğŸ“± Paid â‚¹{amount} using UPI.")

payments = [CreditCard(), PayPal(), UPI()]
print("ğŸ”¸ E-commerce Payment Polymorphism:")
for method in payments:
    method.make_payment(999)

print("\n")
```
